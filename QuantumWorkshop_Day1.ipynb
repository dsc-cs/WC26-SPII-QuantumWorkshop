# Essential Quantum Mechanics

**“I think I can safely say that nobody understands quantum mechanics.”**  
— Richard Feynman (1965)

Quantum mechanics principles are essential for constructing quantum hardware, but they are not the primary focus when learning quantum computing concepts.
- Instead, we’ll focus on simple models and algorithms that help explain how quantum information behaves, without getting into the physics.

While you don’t need deep physics to understand the basics of quantum computers, the following ideas can help explain why quantum computing behave differently from classical ones:

1. **Superposition:** a quantum system can exist in multiple states at the same time (e.g., both 0 and 1). This is what allows quantum computers to process many possibilities simultaneously.

   ▶ [Watch: Superposition Explained](https://www.youtube.com/watch?v=IHDMJqJHCQg)

2. **Measurement:** observing a quantum system forces it to “choose” one outcome from many possibilities. This is why quantum results are inherently uncertain, which is why programs must be run many times.

3. **Entanglement:** Two or more qubits can become linked so that the state of one is related to the state of another. This enables coordinated multi-qubit behavior and is essential for real quantum computation.

   ▶ [Watch: Entanglement Explained](https://www.youtube.com/watch?v=rqmIVeheTVU)

4. **Probability Distributions:** Quantum results follow probability patterns rather than producing a single fixed answer. Results are interpreted statistically, not as one “correct” output.

5. **Noise:** Real quantum systems are affected by environmental interference, which introduces errors. This is why quantum results are imperfect and why error mitigation is an important area of research.

#Quantum Computing Terminology

1. **Quantum Advantage:** The performance benefit quantum computers can achieve over classical computers for certain problem types, such as simulating quantum systems or performing specific searches more efficiently.

2. **Quantum Parallelism:** The ability of a quantum computer to apply one operation to many possible values at the same time, similar to an unlimited SIMD (Single Instruction, Multiple Data) model, enabling faster computation.

![SIMD](https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/simd.png)

  - One operation acts on many possibilities at once, like SIMD.
  - **But** you never see the intermediate values.
  - The computation reshapes a probability distribution using interference:
    - Good answers are amplified
    - Bad answers are suppressed
  - Measurement gives you one sample, not the full state.

3. **Shot:** A single execution of a quantum circuit that produces one  outcome. Because results are probabilistic, many shots are run to estimate the full probability distribution of possible outcomes.

4. **Quantum State:** A mathematical description of a quantum system that encodes all possible outcomes of a computation, along with their probabilities and relative phase (direction) information.

5. **Measurement:** The process of observing a quantum state, which produces one probabilistic outcome from many possibilities and requires repeated runs (shots) to understand the full distribution.

6. **Amplitude:** A number that describes how strongly a quantum state contributes to the overall system, and its square determines the probability of observing that state when measured.

7. **Butterfly Computational Pattern:** A structure of quantum operations where pairs of values are repeatedly recombined in parallel to shape the final probability distribution before measurement.

8. **Qubit (Quantum Bit):** The basic unit of information in a quantum computer. Unlike a classical bit that is either 0 or 1, a qubit can exist in a superposition of both states at the same time, and can also become entangled with other qubits to enable powerful quantum computations.

9. **Quantum Superposition:** The ability of a quantum system to exist in multiple states at the same time, leading to uncertainty in measurement outcomes.

10. **Quantum Entanglement:** A property where qubits become linked so that their states cannot be described independently, making their measurement results interdependent.

11. **Measurement Uncertainty**: The inherent unpredictability of which outcome will be observed when measuring a quantum system in superposition.

##Why Quantum?

Quantum computing offers speedups beyond classical scaling for select problems (e.g., Fourier transforms, search, simulation). This is known as the **quantum advantage**.

![QuantumVsClassical](https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/quantum-vs-classical.png)

Computations that are impractical or impossible classically (quantum physics, chemistry, true randomness) are significantly faster
- e.g., in chemistry, predicting a molecule’s energy levels requires modeling how many interacting electrons behave under quantum mechanics.
  - Each additional electron doubles the size of the system’s state space, causing classical simulations to scale exponentially in time and memory.
- Even powerful classical supercomputers struggle with accurate simulations for moderately sized molecules (beyond a few dozen interacting electrons).
  - A quantum computer represents electron states directly using qubits, allowing it to naturally model superposition and entanglement.
- Quantum algorithms can estimate molecular energy levels with far fewer resources, enabling applications such as drug discovery, catalyst design, and materials science.

![MoleculeCost](https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/molecule-cost.png)




In addition to speedups and reduction of resources, quantum computers can perform some computations that classical computers cannot, such as
- **quantum modeling:** They can simulate molecular behavior at the quantum level, which is useful for applications such as drug discovery and energy technologies.
- **true randomness:** Quantum systems produce inherently random outcomes, unlike classical systems that rely on deterministic algorithms or pseudo-random number generators.
  - This is especially valuable for cryptography and security applications

## Key Characteristics

- Quantum parallelism: single operations act on many value pairs simultaneously (SIMD-like)
- Parallel updates can cause unwanted side effects, requiring careful algorithm design
- Measurement is probabilistic: one run gives one outcome from a distribution
- Repeated runs (“shots”) reveal probability distributions, not deterministic answers
- Results are read as binary strings, enabling digital or probabilistic interpretations

We can demonstrate quantum parallelism by swapping odd and even components of an 8-element array.

For a classical implementation, we can write out all the operations or use a for loop to execute sequentially (Fig 1.2)

     alt="Sequential Swap"

```python
# explicit loop-based swap

values = [0.05, 0.11, 0.13, 0.02, 0.34, 0.17, 0.06, 0.12]

print('before swap:', values)
# swap adjacent elements (0↔1, 2↔3, 4↔5, 6↔7)
for i in range(0, len(values), 2):
    values[i], values[i+1] = values[i+1], values[i]

print('after swap:',values)
```

The code shown above is an example of an algorithm which uses explicit, conventional iteration (i.e., a loop).

We can use Python slicing to demonstrate a more transparent version, which  removes the visible loop.



```python
# slice-based swap

values = [0.05, 0.11, 0.13, 0.02, 0.34, 0.17, 0.06, 0.12]
print('before swap:', values)
values[::2], values[1::2] = values[1::2], values[::2]
print('after swap:',values)
```

This version appears to swap all adjacent elements at once, but the underlying implementation still performs conventional iteration
- It is simply abstracted away in optimized C code.

- The loop has not disappeared; it is just hidden.

Quantum state evolution is fundamentally different.

We treat the state as one mathematical object that gets transformed in a single step.

<img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/quantum-swap.png"
     alt="Quantum Swap"
     width="50%">

Instead of updating individual entries one-by-one, the entire state vector is multiplied by a transformation matrix.
- The operation applies to all possibilities simultaneously because they are part of the same mathematical object.
- In classical computing, even when iteration is hidden, individual elements are still processed independently.
- In quantum computing, there are no independent values being iterated over; there is a single state evolving as a whole.

## Quantum Algorithms

- Programs evolve a quantum state across exponentially many outcomes in parallel (**quantum parallelism**)
  - Instead of checking each possibility one by one, quantum algorithms use simple repeated steps that mix the possibilities together.
- They use linear transformations that recombine values in repeating “butterfly” patterns

  <img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/butterfly-pattern.png"
     alt="Butterfly Pattern"
     width="50%">

- Think of this as stirring ingredients in a bowl; each simple step mixes everything together a little more, gradually increasing the amount of the “right” answer and reducing the rest.
  - After this mixing process shapes the probabilities, measurement produces a single classical outcome (bit, integer, eigenvalue, etc.).
  - The solution is not in a single run, but in the statistical pattern that appears across many runs.

- Note that in the stirring analogy above, we do not know the right answer ahead of time; the algorithm is designed so that the structure of the problem causes the correct answers to amplify automatically.
  - A quantum algorithm doesn’t “guess and boost.” Instead:
  - The problem is encoded into a function (often called an **oracle**).
  - That function marks correct solutions in a mathematically detectable way.
  - The quantum circuit is built so that interference increases the amplitudes of those marked states.
  - We are not manually choosing which answer to amplify; the problem structure determines it.

## Interference does the work

Constructive interference increases amplitudes of states that satisfy the condition, while destructive interference suppresses the others through repeated linear transformations (the butterfly recombinations).

## Example: Grover’s Algorithm
- Suppose one item in a database satisfies a condition, but we do not know which one.
- The algorithm begins in an equal superposition, meaning every possible input starts with the same amplitude (and therefore the same probability).
- A quantum circuit (an oracle) is designed to evaluate the condition f(x).
- When applied to the superposition, it flips the phase of any state that satisfies the condition.
- The function is applied to all possible inputs at once, without checking them individually.
- Repeated mixing steps then use interference to increase the amplitude of the state that satisfies the condition.
- When the system is measured, the correct item appears with high probability.

- Common quantum algorithm patterns:
  - Sampling from hard-to-generate distributions
  - Searching for digital answers (e.g., Grover, Shor)
  - Estimating probabilities via many shots (analog-style results)
  - Emphasis on hardware- and library-agnostic thinking as the ecosystem evolves

### Equal Superposition
- An **equal superposition** is a quantum state where all possible outcomes have equal weight, meaning they have the same probability when measured.
- After qubits are initialized, equal superposition creates a balanced starting point where no outcome is favored and interference can amplify the correct answer.
- For a single qubit, an equal superposition is written as (|0> + |1>) / &radic;2
- In this state:
  - The amplitudes of |0⟩ and |1⟩ are equal.
  - Each outcome has a 50% probability when measured.
  - On a graph, this appears as two bars of equal height.

  <img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/equal-superposition.png"
      alt="Equal Superposition"
      width="50%">
  
  - On the Bloch sphere, this corresponds to any point on the equator.
    - Although all equator points give equal measurement probabilities, they can differ in phase, meaning they are not identical states.
- Equal superposition does not mean the qubit is classically “both 0 and 1”; it means both possibilities are equally weighted before measurement.

### 3. Apply a Simple Quantum Operation
To understand how quantum computation works, we begin with a simple quantum operation called a **gate**.
- A quantum gate changes the state of a qubit, similar to how a logic gate changes a classical bit.
- Gates are the basic building blocks of quantum circuits.
- A **Hadamard gate** (named after French mathematician Jacques Hadamard) is one of the most fundamental quantum gates; it shows us how a qubit can be transformed from a definite state into a superposition.
- The Hadamard gate's main purpose is to create superposition.
- If we start with |0> (ground state), applying the Hadamard gate (H) produces an equal superposition:

  (|0> + |1>) / √2

- This means the qubit is now equally likely to be measured as 0 or 1.

- A simple circuit diagram of this operation looks like this:

  q0: ──|0⟩──[ H ]──[ M ]──▶ 0/1

  where q0 is the name of the qubit.

  |0⟩: The qubit (q0) starts in the initialized ground state.

  [ H ] The Hadamard gate is applied, which transforms q0 into an equal superposition state.

  [ M ] The qubit is measured, which collapses the superposition to either 0 or 1 with roughly 50% probability each.

- **NOTE** a gate applies a mathematical transformation to the entire state at once. It does not modify values one by one — it transforms the full state in a single step.
- **NOTE 2** superposition is not a default state; it is produced by a quantum operation.
- **NOTE 3** on real hardware, a Hadamard gate is implemented by applying a precisely shaped microwave or laser pulse that rotates the qubit's quantum state (changes the direction of its state vector on the Bloch sphere)

### 4. Simulate Measurement
- In the above steps we mathematically transformed a qubit's state using the Hadamard gate.
- Now let's move from abstract state vectors to observable outcomes.
- In this step, we will simulate repeated measurements (“shots”) of the qubit, record how often each result occurs, and plot the distribution using some basic Python libraries.
- You do not need to know Python to follow along, focus on the concepts; the code is simply automating the measurement and counting process for us.
- This allows us to connect probability amplitudes (from the state vector) to real experimental frequencies, demonstrating how quantum theory predicts measurable behavior.

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Define the Hadamard gate
# -----------------------------
H = (1 / np.sqrt(2)) * np.array([
    [1,  1],
    [1, -1]
])

# -----------------------------
# Step 2: Define initial state |0>
# -----------------------------
state = np.array([1, 0], dtype=complex)

print("Initial state |0>:", state)

# -----------------------------
# Step 3: Apply Hadamard (rotation on Bloch sphere)
# -----------------------------
state = H @ state

print("\nState after applying Hadamard:")
print(state)

# -----------------------------
# Step 4: Convert amplitudes to probabilities
# -----------------------------
probabilities = np.abs(state)**2

print("\nMeasurement probabilities:")
print(f"P(0) = {probabilities[0]:.4f}")
print(f"P(1) = {probabilities[1]:.4f}")

# -----------------------------
# Step 5: Simulate repeated measurements (shots)
# -----------------------------
shots = 1000
results = np.random.choice([0, 1], size=shots, p=probabilities)

# -----------------------------
# Step 6: Plot results
# -----------------------------
count_0 = np.sum(results == 0)
count_1 = np.sum(results == 1)

p0 = count_0 / shots
p1 = count_1 / shots
print(f"Frequencies: P(0)≈{p0:.3f}, P(1)≈{p1:.3f}")

plt.bar(["0", "1"], [count_0, count_1])
plt.title("Measurement Results")
plt.ylabel("Counts")
plt.xlabel("Outcome")
plt.show()
```

- In this example, the Hadamard transformed |0⟩ into an equal superposition with amplitudes ≈ 0.7071 and theoretical probabilities 0.5000 / 0.5000.
- Simulating 1000 measurements produced frequencies of approximately 0.517 (for 0) and 0.483 (for 1).
- The slight imbalance reflects normal statistical fluctuation from finite sampling.
- Overall, the experimental frequencies closely match the predicted 50/50 quantum probabilities.

### Qiskit
- Qiskit is an open-source quantum computing SDK for building, simulating, and running quantum circuits on real quantum hardware or simulators.
- The following program recreates the above Hadamard demonstration using Qiskit.
- Instead of manually multiplying matrices, we build a quantum circuit, apply a Hadamard gate, measure the qubit, and run repeated shots on a simulator to observe the resulting distribution.
- This approach mirrors how real quantum hardware is programmed, allowing us to connect conceptual quantum operations to practical circuit execution.

```python
!pip install qiskit qiskit-aer

import numpy as np
import matplotlib.pyplot as plt

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# -----------------------------
# Build circuit: |0> -> H -> measure
# -----------------------------
qc = QuantumCircuit(1, 1)   # 1 qubit, 1 classical bit
qc.h(0)                     # Hadamard gate
qc.measure(0, 0)            # Measure qubit into classical bit

print(qc)

# -----------------------------
# Run "shots" on a simulator
# -----------------------------
shots = 1000
sim = AerSimulator()

compiled = transpile(qc, sim)
job = sim.run(compiled, shots=shots)
result = job.result()
counts = result.get_counts(compiled)

print("\nCounts:", counts)

# -----------------------------
# Plot distribution
# -----------------------------
# Ensure both outcomes appear in a consistent order even if one is missing
count_0 = counts.get("0", 0)
count_1 = counts.get("1", 0)

plt.bar(["0", "1"], [count_0, count_1])
plt.title("Measurement Results After Hadamard (Qiskit Aer)")
plt.ylabel("Counts")
plt.xlabel("Outcome")
plt.show()
```

## What Does a Single Run (Shot) Look Like?

  <img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/knapsack-figure2.2.png"
      alt="Knapsack Problem Figure 2.2"
      width="125%">

A single run of a quantum computation involves preparing the qubits in an initial state, applying a sequence of quantum operations, and then measuring the system to produce one outcome.
- “counts” in this figure refers to the number of times each measurement outcome appears after running the quantum circuit many times.
- By repeating this process many times, we gather a distribution of results.
- **"The goal for many quantum computations is to increase the
probability of measuring desired outcomes."**

### Representing outcomes with binary strings
- There are 2n binary strings with length n, where n > 0 is the number of digits.
- We often talk about measurement outcomes using decimal (base 10 integers)
- The following tables show the decimal (base 10) values
of binary strings of lengths one, two, and three.

  <img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/binary-decimal.png"
      alt="Binary/Decimal Strings"
      width="75%">

- Each qubit in a quantum system corresponds to one binary digit in the measurement outcome.
- A quantum system with n qubits has 2n possible outcomes.

  <img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/qubit-scorecard-2023.png"
      alt="2023 Qubit Scorecard"
      width="75%">
      (QuSecure, n.d.)

**Quantum registers** are groups of qubits used to represent different variables in a computation, similar to variables in classical systems.
- In the knapsack problem, separate registers are used to track item selection, total value, and total weight.

- Three registers encoding selection, value, and weight
  <img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-SPII-QuantumWorkshop/main/images/knapsack-figure2.4.png"
      alt="Three registers for Knapsack Problem"
      width="75%">

  - The number of qubits in each register depends on how much information must be represented for the problem.
  - The selection register uses one qubit per item to represent whether each item is included (1) or not (0).
  - Additional registers store the total value and weight, which are calculated as functions of the binary selection variables.
  - This structured encoding prepares the problem for implementation within a quantum circuit.

The knapsack problem can be solved using a solve_knapsack function that takes item values, weights, and a maximum weight limit as inputs.
- The algorithm uses two key quantum patterns: an oracle to recognize valid selections and amplitude amplification to increase the probability of desirable outcomes.
- It begins by identifying selections that meet the weight constraint and a minimum value threshold, then iteratively raises the minimum value while amplifying valid solutions.
- Through repeated measurement and refinement, the algorithm ultimately returns the optimal selection that maximizes value without exceeding the weight limit.


## Why Do Quantum Computers Look So Weird?
(Post-Quantum, n.d.)

- Quantum computers resemble chandeliers because their design supports the extreme cooling and shielding required for superconducting qubits.
- Cryogenic cooling to millikelvin temperatures (20–100 mK) is essential for superconducting qubits; even tiny amounts of thermal energy cause decoherence (the loss quantum state from environmental disturbance) and computational errors.
- At near-absolute-zero temperatures, materials become superconducting with zero electrical resistance, which enables key quantum computing techniques and preserves coherence.
- Suspending quantum computers from the ceiling reduces ground-borne mechanical vibrations that could disturb fragile qubit states, creating a more stable operating environment.
- A "chandelier-style" vertical design improves thermal isolation, maintains ultra-low temperatures, and allows efficient organization of cryogenic stages and wiring for easier maintenance.
